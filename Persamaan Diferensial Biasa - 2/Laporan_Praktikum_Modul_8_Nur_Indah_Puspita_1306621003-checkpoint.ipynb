# 1. Problem Statement

 Menyelesaikan persamaan diferensial biasa (ODE) dengan metode numerik seperti Metode Runge-Kutta orde 2, 3, 4, dan Metode Runge-Kutta Fehlberg adaptif,  melakukan analisis perbandingan dengan membuat plot grafik solusi menggunakan masing-masing metode, membuat plot grafik galat untuk membandingkan tingkat akurasi antara metode-metode tersebut, dan membuat plot grafik fluktuatif untuk mendapatkan akurasi yang optimal.

# 2. Mathematical Equation

### Metode Runge-Kutta Orde:<br>
1. Runge-Kutta orde 2 <br>
$k1 = f(x_{i}, y_{i})$<br>
$k2 = f(x_{i} + \frac{1}{2} h, y_{i} + \frac{1}{2} hk_{1})$<br>
Solusi akhir :<br>
$y_{i+1} = y_{i} + k_2{h}$

2. Runge-Kutta orde 3 <br>
$k1 = f(x_{i}, y_{i})$<br>
$k2 = f(x_{i} + \frac{1}{2} h, y_{i} + \frac{1}{2} hk_{1})$<br>
$k3 = f(x_{i}+h, y_{i} - k_{1}h + 2k_{2}h)$<br>
Solusi akhir :<br>
$y_{i+1} = y_{i} + \frac{1}{6} (k_{1} + 4k_{2} + k_{3})h$

3. Runge-Kutta orde 4 <br>
$k1 = f(x_{i}, y_{i})$<br>
$k2 = f(x_{i} + \frac{1}{2} h, y_{i} + \frac{1}{2} hk_{1})$<br>
$k3 = f(x_{i} + \frac{1}{2} h, y_{i} + \frac{1}{2} hk_{2})$<br>
$k4 = f(x_{i} + h, y_{i} + k_{3}h)$<br>
Solusi akhir :<br>
$y_{i+1} = y_{i} + \frac{1}{6} (k_{1} + 2k_{2} + 2k_{3} + k_{4})h$

### Metode Runge-Kutta Fehlberg:<br>
$k1 = hf(x_{i}, y_{i})$<br>
$k2 = hf(x_{i} + \frac{h}{4}, y_{i} + \frac{k1}{4})$<br>
$k3 = hf(x_{i} + \frac{3h}{8}, y_{i} + \frac{3}{32}k1 + \frac{9}{32}k2)$<br>
$k4 = hf(x_{i} + \frac{12h}{13}, y_{i} + \frac{1932}{2197}k1 - \frac{7200}{2197}k2 + \frac{7296}{2197}k3)$<br>
$k5 = hf(x_{i} + h, y_{i} + \frac{439}{216}k1 - 8k2 + \frac{3680}{513}k3 - \frac{845}{4104}k4)$<br>
$k6 = hf(x_{i} + \frac{h}{2}, y_{i} - \frac{8}{27}k1 + 2k2 - \frac{3544}{2565}k3 + \frac{1859}{4104}k4 - \frac{11}{40}k5)$<br><br>
Solusi akhir :<br>
$y_{i+1} = y_{i} + \frac{25}{216}k_{1} + \frac{1408}{2565}k_{3} + \frac{2197}{4104}k_{4} - \frac{1}{5}k_{5}$<br>
$\hat{y}_{i+1} = y_{i} + \frac{16}{135}k_{1} + \frac{6656}{12825}k_{3} + \frac{28561}{56430}k_{4} - \frac{9}{50}k_{5} + \frac{2}{55}k_{6}$<br><br>
Mencari error local :<br>
$\epsilon = \hat{y}_{i+1} - y_{i+1}$<br><br>
Ukuran langkah diperbarui dengan ketentuan :<br>
$h_{i+1} = \left\{\begin{matrix}
h_{i}, \epsilon \leq \tau  & \\ 
 \delta h_{i}, \epsilon > \tau& 
\end{matrix}\right.$ <br><br>
Dengan $\delta$ :<br>
$\delta = 0.84089(h_{i} \frac{\tau}{\epsilon})^{\frac{1}{4}}$

```python
import numpy as np
import math
import matplotlib.pyplot as plt

x0 = 0
xend = 35
y0 = 2
step = [0.5, 1, 1.5, 2]

def fungsi(x,y):
    f = (-y) + (0.1 * np.cos(1.5*x) * np.exp(x/4)) 
    return f

def runge2(fungsi, x0, xend, y0, stepsize):
    ynew = [y0]
    xnew = np.arange(x0, xend, stepsize)
    for index, i in enumerate (xnew):
        if index == len(xnew)-1 :
            break
        k1 = fungsi(i, ynew[index])
        k2 = fungsi(i + (stepsize/2), ynew[index] + (stepsize*k1/2))
        ynew.append(ynew[index] + k2*stepsize)
    return xnew, ynew
    
def runge3(fungsi, x0, xend, y0, stepsize):
    ynew = [y0]
    xnew = np.arange(x0, xend, stepsize)
    for index, i in enumerate (xnew):
        if index == len(xnew)-1 :
            break
        k1 = fungsi(i, ynew[index])
        k2 = fungsi(i + (stepsize/2), ynew[index] + (stepsize*k1/2))
        k3 = fungsi(i + stepsize, ynew[index] - k1*stepsize + 2*k2*stepsize)
        ynew.append(ynew[index] + stepsize/6*(k1 + 4*k2 + k3))
    return xnew, ynew

def runge4(fungsi, x0, xend, y0, stepsize):
    ynew = [y0]
    xnew = np.arange(x0, xend, stepsize)
    for index, i in enumerate (xnew):
        if index == len(xnew)-1 :
            break
        k1 = fungsi(i, ynew[index])
        k2 = fungsi(i + (stepsize/2), ynew[index] + (stepsize*k1/2))
        k3 = fungsi(i + (stepsize/2), ynew[index] + (stepsize*k2/2))
        k4 = fungsi(i + stepsize, ynew[index] + k3*stepsize)
        ynew.append(ynew[index] + stepsize/6*(k1 + 2*k2 + 2*k3 + k4))
    return xnew, ynew

def eksak (x):
    f = (2/61 * np.cos(3/2 * x) * np.exp(x/4)) + (12/305 * np.sin(3/2 * x) * np.exp(x/4)) + (120/61 * np.exp(-x))
    return f

#Nomer 1
print('{0:〜^63}'.format('〜'))
print('{0:^114}'.format('Nomer 1 & Nomer 2'))
print('{0:〜^63}'.format('〜'))

for index, i in enumerate (step):
    xrunge2, yrunge2 = runge2(fungsi, x0, xend, y0, i)
    xrunge3, yrunge3 = runge3(fungsi, x0, xend, y0, i)
    xrunge4, yrunge4 = runge4(fungsi, x0, xend, y0, i)
    yeksak = eksak(xrunge2)
    galatrunge2 = abs((np.array(yeksak) - np.array(yrunge2)))
    galatrunge3 = abs((np.array(yeksak) - np.array(yrunge3)))
    galatrunge4 = abs((np.array(yeksak) - np.array(yrunge4)))
    print('\n {0:^114}  '.format(f' Solusi Runge-Kutta (Δx = {i})'))
    print('{0:=^114}'.format('='))
    print('| {0:<5} | {1:^12} | {2:^12} | {3:^12} | {4:^12} | {5:^12} | {6:^12} | {7:^12} |'.format('x', 'RK-Orde 2', 'RK-Orde 3', 'RK-Orde4', 'yeksak', 'Galat RK-2', 'Galat RK-3','Galat RK-4'))
    print('{0:=^114}'.format('='))
    for i in range (len(xrunge2)):
        print('| {0:<5} | {1:^12} | {2:^12} | {3:^12} | {4:^12} | {5:^12} | {6:^12} | {7:^12} |'.format(xrunge2[i], round(yrunge2[i],7), round(yrunge3[i],7), round(yrunge4[i],7), round(yeksak[i],7), round(galatrunge2[i],7), round(galatrunge3[i],7),round(galatrunge4[i],7)))
    
    print('{0:=^114}'.format('='))
    
for index, i in enumerate (step):
    xrunge2, yrunge2 = runge2(fungsi, x0, xend, y0, i)
    xrunge3, yrunge3 = runge3(fungsi, x0, xend, y0, i)
    xrunge4, yrunge4 = runge4(fungsi, x0, xend, y0, i)
    xnew = np.arange(x0, xrunge2[-1], 0.001)
    eksakrunge = eksak(xnew)
    
    plt.figure(figsize = (15,6))
    plt.scatter(xrunge2, yrunge2, label="Metode Runge-Kutta Orde 2", color="purple")
    plt.scatter(xrunge3, yrunge3, label="Metode Runge-Kutta Orde 3", color="green")
    plt.scatter(xrunge4, yrunge4, label="Metode Runge-Kutta Orde 4", color="blue")
    plt.plot(xnew, eksakrunge, label="Metode Eksak", color="red")
    plt.title(f"Hasil Numerik dan Eksak Metode Runge-Kutta (Δx = {i})")
    plt.xlabel("x")
    plt.ylabel("y")
    plt.legend()
    plt.show()
    
```

```python
#Nomer 3
print('{0:〜^65}'.format('〜'))
print('{0:^119}'.format('Nomer 3'))
print('{0:〜^65}'.format('〜'))

x0 = 0
xend = 35
y0 = 2
step = 2
toleransi = [0.1, 0.01, 0.001, 0.0001]

def fungsi(x,y):
    f = (-y) + (0.1 * np.cos(1.5*x) * np.exp(x/4)) 
    return f

def rkf(fungsi, y0, x0, xend, stepsize, toleransi, eksak):
    yrka = [y0]
    yrka1 = [y0]
    neksak = []
    delta = [0]
    h = [stepsize]
    x = [x0]
    n = 0
    
    while x[-1] < xend :
        k1 = h[n]*fungsi(x[n], yrka[n])
        k2 = h[n]*fungsi(x[n] + h[n]/4, yrka[n] + k1/4)
        k3 = h[n]*fungsi(x[n] + (3*h[n]/8), yrka[n] + (3*k1/32) + (9*k2/32))
        k4 = h[n]*fungsi(x[n] + (12*h[n]/13), yrka[n] + (1932*k1/2197) - (7200*k2/2197) + (7296*k3/2197))
        k5 = h[n]*fungsi(x[n] + h[n], yrka[n] + (439*k1/216) - 8*k2 + (3680*k3/513) - (845*k4/4104))
        k6 = h[n]*fungsi(x[n] + (h[n]/2), yrka[n] - (8*k1/27) + 2*k2 - (3544*k3/2565) + (1859*k4/4104) - (11*k5/40))
        
        yrka.append(yrka[n] + (25*k1/216) + (1408*k3/2565) + (2197*k4/4104) - (k5/5))
        yrka1.append(yrka[n] + (16*k1/135) + (6656*k3/12825) + (28561*k4/56430) - (9*k5/50) + (2*k6/55))
        error = abs(yrka1[n] - yrka[n])
        
        if error <= toleransi:
            h.append(stepsize)
            delta.append(delta[-1])
        elif error > toleransi:
            delta.append(0.84089 * (h[n] * toleransi / error) ** (1/4))
            h.append(h[n] * 0.84089 * (h[n] * toleransi / error) ** (1/4) )
            
        x.append(x[-1] + h[n])
        neksak.append(eksak(x[n]))
        
        n += 1
    return x, yrka, yrka1, delta, h, n, neksak

def eksak (x):
    f = (2/61 * np.cos(3/2 * x) * np.exp(x/4)) + (12/305 * np.sin(3/2 * x) * np.exp(x/4)) + (120/61 * np.exp(-x))
    return f

#Mencetak
for index, i in enumerate (toleransi):
    xrkf, yrka, yrka1, delta, steph, n, neksak = rkf(fungsi, y0, x0, xend, step, i, eksak)
    
    print('\n {0:^119}  '.format(f' Solusi Runge-Kutta Fehlberg (Toleransi = {i})'))
    print('{0:=^119}'.format('='))
    print('| {0:<5} | {1:^12} | {2:^22} | {3:^22} | {4:^12} | {5:^12} | {6:^12} |'.format('No', 'X', 'Y Runge-Kutta Orde 4', 'Y Runge-Kutta Orde 5', 'Y Eksak', 'Step (h)', ' \u03B4 (Delta)'))
    print('{0:=^119}'.format('='))
    
    no = 0
    
    for i in range(len(xrkf)-1):
        print('| {0:<5} | {1:^12} | {2:^22} | {3:^22} | {4:^12} | {5:^12} | {6:^12} |'.format(no, round(xrkf[i], 7), yrka[i], yrka1[i], round(neksak[i], 7), round(steph[i], 7), round(delta[i], 8)))
        no += 1
    print('{0:=^119}'.format('='))
    
for index, i in enumerate (toleransi):
    xrkf, yrka, yrka1, delta, steph, n, neksak = rkf(fungsi, y0, x0, xend, step, i, eksak)
    xnew = np.arange(x0, xrkf[-1], 0.001)
    eksakrkf = eksak(xnew)
    
    plt.figure(figsize = (15,6))
    plt.scatter(xrkf, yrka1, label="Metode Runge-Kutta Fehlberg", color="purple")
    plt.plot(xnew, eksakrkf, label="Metode Eksak", color="blue")
    plt.title(f"Hasil Numerik dan Eksak Metode Runge-Kutta Fehlberg(Toleransi = {i})")
    plt.xlabel("x")
    plt.ylabel("y")
    plt.legend()
    plt.show()
```
