### 2. Mathematical Equation

### **Persamaan Umum :**<br>
$y = a_{0} + a_{1}x_{1} + a_{2}x_{2} + .... + a_{n}x_{n}$<br><br>

**Persamaan regresi linear orde - 1 :**<br>
Persamaan 1 : $\left ( \sum y_{i} \right ) \space\space\space= na_{0} + \left ( \sum x_{i} \right )a_{1}$<br>

Persamaan 2 : $\left ( \sum y_{i}x_{i} \right ) = \left ( \sum x_{i} \right )a_{0} + \left ( \sum x_{i} ^{2} \right )a_{1}$<br><br>

**Persamaan regresi linear orde - 2 :**<br>
Persamaan 1 : $\left ( \sum y_{i} \right ) \space\space\space = na_{0}\space+\space\left ( \sum x_{i} \right )a_{1} + \left ( \sum x_{i} ^{2} \right )a_{2}$<br>

Persamaan 2 : $\left ( \sum y_{i}x_{i} \right ) = \left ( \sum x_{i} \right )a_{0} + \left ( \sum x_{i} ^{2} \right )a_{1} + \left ( \sum x_{i} ^{3} \right )a_{2}$<br>

Persamaan 3 : $\left ( \sum y_{i}x_{i}^{2} \right ) = \left ( \sum x_{i}^{2} \right )a_{0} + \left ( \sum x_{i} ^{3} \right )a_{1} + \left ( \sum x_{i} ^{4} \right )a_{2}$<br><br>

**Persamaan regresi linear orde - 3 :**<br>
Persamaan 1 : $\left ( \sum y_{i} \right ) \space\space\space = na_{0}\space+\space\left ( \sum x_{i} \right )a_{1} + \left ( \sum x_{i} ^{2} \right )a_{2} + \left ( \sum x_{i} ^{3} \right )a_{3}$<br>

Persamaan 2 : $\left ( \sum y_{i}x_{i} \right ) = \left ( \sum x_{i} \right )a_{0} + \left ( \sum x_{i} ^{2} \right )a_{1} + \left ( \sum x_{i} ^{3} \right )a_{2} + \left ( \sum x_{i} ^{4} \right )a_{3}$<br>

Persamaan 3 : $\left ( \sum y_{i}x_{i}^{2} \right ) = \left ( \sum x_{i}^{2} \right )a_{0} + \left ( \sum x_{i} ^{3} \right )a_{1} + \left ( \sum x_{i} ^{4} \right )a_{2} + \left ( \sum x_{i} ^{5} \right )a_{3}$<br>

**Metode Eliminasi Gauss Untuk Mencari nilai $a_{n}$**

    - $a_{jk} - m_{jk}a_{kk} = a_{jk} - \frac{a_{jk}}{a_{kk}} a_{kk}$<br>
2. Substitusi Mundur :<br>
    - $x_{i} = \frac{1}{a_{ii}}(d_{i}-\sum_{n}^{j} a_{ij}x_{j})$

### 5. Source Code

### **Soal a**

```python
import numpy as np
import matplotlib.pyplot as plt

# Data
X = np.array([0.5, 1.43, 2.86, 4.29, 5.71, 7.14, 8.57, 11.42, 12.86, 14.29, 15.71, 17.14, 18.57, 20.00])
Y = np.array([40.226, 103.33, 134.14, 134.25, 107.57, 112.66, 108.78, 130.05, 126.37, 112.81, 120.29, 110.44, 116.30, 118.56])

# Regresi polinomial derajat 1
z1 = np.polyfit(X, Y, 1)
p1 = np.poly1d(z1)
print("Maka persamaan regresi linear orde 1 adalah\n", p1)

# Regresi polinomial derajat 2
z2 = np.polyfit(X, Y, 2)
p2 = np.poly1d(z2)
print("Maka persamaan regresi linear orde 2 adalah\n", p2)

# Regresi polinomial derajat 3
z3 = np.polyfit(X, Y, 3)
p3 = np.poly1d(z3)
print("Maka persamaan regresi linear orde 3 adalah\n", p3)
```

### **Soal b**

```python
# Menghitung SSE dan SST
y_mean = np.mean(p3(X))
SSE1 = np.sum((Y - p1(X))**2)
SSE2 = np.sum((Y - p2(X))**2)
SSE3 = np.sum((Y - p3(X))**2)
SST = np.sum((1/n)*(Y - y_mean)**2)

# Menghitung standar deviasi dari tiga fungsi aproksimasi
n = len(X)
sd1 = np.sqrt(SSE1 / (n - 2))
sd2 = np.sqrt(SSE2 / (n - 3))
sd3 = np.sqrt(SSE3 / (n - 4))

# Menampilkan SSE, SST, dan standar deviasi dari tiga fungsi aproksimasi
print("SSE (derajat 1) = ", SSE1)
print("SSE (derajat 2) = ", SSE2)
print("SSE (derajat 3) = ", SSE3)
print("SST = ", SST)
print("Standar deviasi (derajat 1) = ", sd1)
print("Standar deviasi (derajat 2) = ", sd2)
print("Standar deviasi (derajat 3) = ", sd3)
```

### **Soal c**

```python
# Plot data dan tiga fungsi aproksimasi
plt.scatter(X, Y, color='blue', label = 'Nilai sebenarnya')
plt.plot(X, p1(X), color='red', label='Regresi polinomial derajat 1')
plt.plot(X, p2(X), color='green', label='Regresi polinomial derajat 2')
plt.plot(X, p3(X), color='purple', label='Regresi polinomial derajat 3')
plt.legend()
plt.show()
```
