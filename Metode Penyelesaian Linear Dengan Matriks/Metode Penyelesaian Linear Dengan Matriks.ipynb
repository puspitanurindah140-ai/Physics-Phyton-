## Gauss Naive 

Rumus Maju :<br><br>
$a_{jk} - m_{jk}a_{kk} = a_{jk} - \frac{a_{jk}}{a_{kk}} a_{kk}$<br><br>

Rumus Mundur :<br><br>
$x_{i} = \frac{1}{a_{ii}}(d_{i}-\sum_{n}^{j} a_{ij}x_{j})$<br><br>

## Gauss Jordan
Rumus: <br><br>
$a_{ij} = a_{ij} - m\times a_{kj} = a_{ij} - a_{ik}\times a_{kj}$<br><br>
$a_{k} = \frac{a_{k}}{a_{kk}}$

### 8. Source Code
**Metode Gauss Jordan**

```python
print ("Nama: Nur Indah Puspita")
print ("NIM : 1306621003")
print ("Metode Gauss Jordan")

import numpy as np

#Diketahui
vo = 12
rs = 0.2
r1 = 220
r2 = 110
r3 = 330
rx = 150
ra = 100

A = np.array([[0.2,220,100,12],
              [-150,470,-100,0],
              [-330,-100,540,0],])
n = len(A)
print("Matriks Augmented\n", A)

n = len(A)
for k in range(n) : 
    pivot = A[k][k]
    A[k] = A[k]/pivot
    for i in range(n):
        if i == k:
            continue
        factor = A[i][k]
        for j in range (k,n+1):
            A[i][j] = A[i][j] - factor*A[k][j]
            
print("Matriks Eselon\n", A)

x = A[:, n]

print("Maka nilai i1, i2, dan i3 dengan metode Gauss Jordan adalah\n", x)
```
