 <font color = pink> 1. Problem Statement </font><br>

Menyelesaikan persamaan diferensial menggunakan metode Euler dan metode Heun, untuk menghasilkan grafik dari solusi numerik menggunakan kedua metode tersebut serta membandingkannya dengan solusi eksak, dan menghitung rata-rata galat antara solusi numerik dan solusi eksak, untuk memperoleh akurasi serta efektivitas kedua metode tersebut.

 <font color = pink> 2. Mathematical Equation </font><br>

### Metode Euler :<br>
$y_{i+1} = y_{i} + f(x_{i}, y_{i})\Delta x$ <br><br>
dimana : $\Delta x = \frac{x_{i+1} - x_{i}}{step}$

### Metode Heun : <br>
Prediktor : $y_{i+1}^{0} = y_{i} + f(x_{i}, y_{i}) \Delta x$ <br><br>
Korektor : $y_{i+1} = y_{i} + \frac{f(x_{i}, y_{i}) + f(x_{i+1}, y_{i+1}^{0})}{2} \Delta x$

```python
print ("Nama: Nur Indah Puspita")
print ("NIM : 1306621003")
print ("Solusi Persamaan Diferensial Biasa dengan Metode Euler dan Heun")
print('-----------------------------------------------------------')

import numpy as np
import matplotlib.pyplot as plt
from colorama import Fore, Style

#diketahui
x0 = 0
xend = 8
y0 = 0.5
step = [0.05, 0.1, 0.25, 0.5, 0.7]

#def
def fungsi(x,y):
    f = (2-x*y)*y
    return f

def euler(fungsi, y0, x0, xend, stepsize):
    ynew = [y0]
    xnew = np.arange(x0, xend+stepsize, stepsize)
    for index, i in enumerate (xnew):
        if index == len(xnew)-1 :
            break
        ynew.append(ynew[index] + stepsize*fungsi(i, ynew[index]))
    return xnew, ynew

def heun(fungsi, y0, x0, xend, stepsize):
    ynew = [y0]
    xnew = np.arange(x0, xend+stepsize, stepsize)
    for index, i in enumerate(xnew):
        if index == len(xnew)-1 :
            break
        ynew.append(ynew[index] + (stepsize/2)*(fungsi(i, ynew[index]) + fungsi(xnew[index + 1], ynew[index] + stepsize*fungsi(i, ynew[index]))))
    return xnew, ynew

def eksak(x):
    f = 4 / (-1 + 9*np.exp(-2*x) + 2*x)
    return f

#Nomer 1
print(Fore.GREEN + '********** NOMOR 1 **********')
print(Style.RESET_ALL)

for index, i in enumerate (step):
    xeuler, yeuler = euler(fungsi, y0, x0, xend, i)
    xheun, yheun = heun(fungsi, y0, x0, xend, i)
    yeksak = eksak(xeuler)
    galateuler = abs((np.array(yeksak) - np.array(yeuler)))
    galatheun = abs((np.array(yeksak) - np.array(yheun)))
    print('{0:^44}        {1:^44}'.format(f'Solusi untuk metode Euler (Δx = {i})', f'Solusi untuk metode Heun (Δx = {i})'))
    print('{0:=^44}        {1:=^43}'.format('=', '='))
    print('| {0:<5} | {1:^8} | {2:^8} | {3:^8} |        | {4:<5} | {5:^8} | {6:^8} | {7:^8} |'.format('x','yeuler','yeksak','galateuler','x','yheun','yeksak','galatheun'))
    print('{0:=^44}        {1:=^43}'.format('=', '='))

    for i in range(len(xeuler)):
        print('| {0:<5} | {1:^8} | {2:^8} | {3:^10} |        | {4:<5} | {5:^8} | {6:^8} | {7:^9} |' .format(round(xeuler[i],5), round(yeuler[i],5), round(yeksak[i],5), round(galateuler[i],5), round(xheun[i],5), round(yheun[i],5), round(yeksak[i],5), round(galatheun[i],5)))

    print('{0:=^44}        {1:=^43} \n'.format('=', '='))
```

```python
#Nomer 2
print(Fore.GREEN + '********** NOMOR 2 **********')
print(Style.RESET_ALL)

print('{0:=^62}'.format('='))
print('| {0:^8} | {1:^21} | {2:^23} |'.format('x', 'Mean Galat Euler', 'Mean Galat Heun'))
print('{0:=^62}'.format('='))

meaneuler = []
meanheun = []
for index, i in enumerate (step):
    xeuler, yeuler = euler(fungsi, y0, x0, xend, i)
    xheun, yheun = heun(fungsi, y0, x0, xend, i)
    yeksak = eksak(xeuler)
    galateuler = abs((np.array(yeksak) - np.array(yeuler)))
    galatheun = abs((np.array(yeksak) - np.array(yheun)))
    meaneuler.append(np.mean(galateuler))
    meanheun.append(np.mean(galatheun))

    print('| {0:^8} | {1:^21} | {2:^23} |'.format(step[index], meaneuler[index], meanheun[index]))

print('{0:=^62}'.format('='))

plt.plot(step, meaneuler, '-o', color = 'green',label = 'Euler')
plt.plot(step, meanheun, '-o',color = 'purple', label = 'Heun')
plt.title('Grafik Perbandingan Rata-Rata Galat dengan Ukuran Langkah Berbeda')
plt.xlabel('Ukuran Langkah (\u0394 x)')
plt.ylabel('Rata-rata Galat')
plt.legend()
```
