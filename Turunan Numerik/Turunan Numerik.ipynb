# 2. Mathematical Equation

Persamaan turunan secara umum :<br><br>
$f'(x_{i}) = \frac{f(x_{i+1}) - f_{x_{i}}}{h} + O(h)$ <br><br>

Metode selisih maju :<br><br>
$f'(x_{i}) = \frac{f(x_{i+1}) - f_{x_{i}}}{h}$ <br><br>

Metode selisih mundur:<br><br>
$f'(x_{i}) = \frac{f(x_{i}) - f_{x_{i-1}}}{h}$ <br><br>

Metode selisih pusat:<br><br>
$f'(x_{i}) = \frac{f(x_{i+1}) - f_{x_{i-1}}}{h}$ <br><br>
