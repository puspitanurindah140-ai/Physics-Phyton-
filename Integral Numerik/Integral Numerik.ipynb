# 1. Problem Statement

Menyesaikan permasalahan pemrograman perhitungan integral menggunakan metode numerik dan menghitung periode osilasi bandul dengan menggunakan metode trapezoidal dan metode simpson pada bidang vertikal. 

# 2. Mathematical Equation

### Definite integral: <br>
a. Analytical calculation: $f(x) = \int_{2}^{3} \ln(x) , dx$ <br>
b. Trapezoidal rule: $I_{\text{trapezoidal}} = \frac{h}{2} \left(f(a) + 2 \sum_{i=1}^{n-1} f(x_i) + f(b)\right)$ <br>
c. Simpson's rule: $I_{\text{Simpson}} = \frac{h}{3} \left(f(a) + 4 \sum_{i=1}^{\frac{n}{2}} f(x_{2i-1}) + 2 \sum_{i=1}^{\frac{n}{2}-1} f(x_{2i}) + f(b)\right)$

###Period of oscillating pendulum: <br>
a. Integral equation: $T = 4\sqrt{\frac{l}{2g}} \int_{0}^{\theta_0} \frac{d\theta}{\sqrt{\cos(\theta) - \cos(\theta_0)}}$ <br>
b. Small-angle approximation: $T_{\text{approx}} = 2\pi\sqrt{\frac{l}{g}}$

### Source Code

```python
print ("Nama: Nur Indah Puspita")
print ("NIM : 1306621003")
print ("Integral Numerik") 
print('-----------------------------------------------------------')

import numpy as np

a = 2
b = 3
n = 1000
h = (b-a)/n

def f(x):
    f = np.log(x)
    return f

def trapesium() :
    s = f(a) + f(b)
    for i in range (1,n):
        s += 2*(f(a+i*h))
    luas = s * h/2
    
    return luas

def simpson() :
    s = f(a) + f(b) 
    for i in range (1,n):
        if i % 2 == 0 :
            c = 2*(f(a+i*h))
        else :
            c = 4*(f(a+i*h))
        s += c
    luas = s * h/3
    
    return luas
    
print(f"Nilai integral dengan metode trapezoidal adalah = {trapesium()}")
print(f"Nilai integral dengan metode simpson adalah = {simpson()}")

#Grafik
import matplotlib.pyplot as plt

xnew = np.linspace(2,3,num=1000)
y = np.log(xnew)

plt.plot(xnew,y, color = 'purple', label ='f(x) = ln(x)')
plt.title('Grafik Integral f(x) = ln(x)')
plt.fill_between(xnew,y, y2=0, color = 'purple', alpha = 0.2, label = 'Integral Trapezoidal')
plt.fill_between(xnew,y, y2=0, color = 'pink', alpha = 0.2, label = 'Integral Simpson')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.show
```

### Source Code

```python
print ("Nama: Nur Indah Puspita")
print ("NIM : 1306621003")
print ("Integral Numerik") 
print('-----------------------------------------------------------')

import numpy as np
import math
from scipy.integrate import quad

#Diketahui
teta = np.array([np.pi/128, np.pi/64, np.pi/32, np.pi/16, np.pi/8, np.pi/4, np.pi/2])

def eksak():
    eksak = 2 * np.pi
    return eksak

def f(x, teta):
    f = (4/np.sqrt(2)) / (np.sqrt((-np.cos(teta)) + (np.cos(x))))
    return f

def analitik (teta) :
    
    pers = lambda x: 4/np.sqrt(2) / math.sqrt(math.cos(x) - math.cos(teta))
    analitik, none = quad(pers, 0, teta)
    
    hasil = analitik
    return hasil
    
def trapesium (teta):
    a = 0
    b = teta
    n = 1000
    h = (b-a) / n
    s = (f(a,b) + f(b-(1/n), b))
    
    for i in range(1,n):
        s += 2*(f(a+i*h, b))
        
    luas = s*(h/2)
    return luas

def simpson(teta) :
    a = 0
    b = teta
    n = 1000
    h = (b-a) / n
    s = (f(a,b) + f(b-(1/n), b))
    
    for i in range (1,n):
        if i % 2 == 0 :
            c = 2*(f(a+i*h, b))
        else :
            c = 4*(f(a+i*h, b))
        s += c
    luas = s * h/3
    
    return luas

def error_trapesium (teta):
    error = abs(((eksak() - trapesium(teta)) / eksak()) * 100)
    return error

def error_simpson (teta):
    error = abs(((eksak() - simpson(teta)) / eksak()) * 100)
    return error

#Tabel
print('Tabel Hasil Perhitungan')
print('-----------------------------------------------------------------------------------------------------------------------------------------------------------------------')
print('| {0:^5} | {1:^15} | {2:^25} | {3:^25} | {4:^25} | {5:^25} | {6:^25} |'.format('No', '\u03F4 0','Perhitungan Analitik', 'Perhitungan Trapezoidal', 'Perhitungan Simpsons', 'Error Trapesium', 'Error Simpson'))
print('-----------------------------------------------------------------------------------------------------------------------------------------------------------------------')
no = 0

for i in range(len(teta)) :
    no += 1
    print('| {0:^5} | {1:^15} | {2:^25} | {3:^25} | {4:^25} | {5:^25} | {6:^25} |'.format(no, round(teta[i], 5), f"{round(analitik(teta[i]), 7)} x sqrt (l/g)", f"{round(trapesium(teta[i]),7)} x sqrt (l/g)", f"{round(simpson(teta[i]),7)} x sqrt (l/g)", error_trapesium(teta[i]), error_simpson(teta[i])))
    
print('-----------------------------------------------------------------------------------------------------------------------------------------------------------------------')

print(f'\n Perhitungan dengan pendekatan sudut terkecil {eksak()} x sqrt (l/g)')

#Grafik
import matplotlib.pyplot as plt

x = []
y = []
y1 = []
analitiks = []
real = eksak()
for i in range(len(teta)):
    x.append(teta[i])
    y.append(trapesium(teta[i]))
    y1.append(simpson(teta[i]))
    analitiks.append(analitik(teta[i]))
   
plt.figure(figsize=(8,5))
plt.title('Grafik Perbandingan Periode dengan sudut \u03F4')
plt.plot(x, y, label='Integral Trapezoidal', color="yellow")
plt.plot(x, y1, label = 'Integral Simpson', color="green")
plt.plot(x, analitiks, label = 'Analitik', color="purple")
plt.axhline(y=real, color='r', linestyle='--', label="Nilai Sudut Terkecil")
plt.legend()
plt.xlabel('\u03F4 (rad)')
plt.ylabel('T (sqrt (l/g))')
plt.show
```
