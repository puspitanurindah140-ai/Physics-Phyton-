### 2. Persamaaan Matematika

Persamaan rapat energi sebagai berikut :<br><br>
$U(\lambda, 5000K) = \frac{8\pi hc}{\lambda} \frac{1}{e^{hc/\lambda K_{B}T} - 1}$<br><br>
Dengan nilai-nilai yang diketahui : <br><br>
$h = 6.626 \times 10^{-34} m^{2}kg/s$<br>
$c = 2.998 \times 10^{8} m/s$<br>
$k_{B} = 1.381 \times 10^{-23} m^{2}kg s^{-2}K^{-1}$<br><br>
Rapat energi matahari dari eksperimen :<br>

|:-------:|:------:|

|   721   | 127919 |
|   972   | 134231 |
|   1020  | 286420 |

|   1726  | 482845 |

Rumus Interpolasi : <br><br>
$P_{n}(x) = \sum_{i=0}^{n}f(xi)L_{i}(x)$

$f(xi) = y_{n}$ dan $L_{i}(x) = \prod_{j=0}^{n} \frac{x - x_{j}}{x_{i} - x{j}}$

Rumus Interpolasi :<br><br>
$f_{n}(x)=b_{0}+b_{1}(x-x_{0})+....+b_{n}(x-x_{1})...(x-x_{n-1})$<br><br>
Dimana : <br><br>
$b_{0}=f(x_{0})=y_{0}$<br>
$b_{1}=f[x_{1},x_{0}]=\frac{f(x_{1})-f(x_{0})}{x_{1}-x_{0}}$<br>
$b_{2}=f[x_{2},x_{1},x_{0}]=\frac{f[x_{2},x_{1}] - f[x_{1},x_{0}]}{x_{2} - x_{0}}$<br><br>

### 5. Source Code

**Metode Interpolasi Lagrange**

```python
print ("Nama: Nur Indah Puspita")
print ("NIM : 1306621003")
print ("Metode Interpolasi Lagrange")

import numpy as np
import matplotlib.pyplot as plt
import math

#Diketahui
h = 6.626*10**-34
c = 2.998*10**8
kb = 1.381*10**-23
pi = np.pi
T = 5000
e = np.e
no = 0

#Nilai x dan y
x = [297, 721, 972, 1020, 1434, 1726] #nilai lamda
y = [75865, 127919, 134231, 286420, 314502, 482845] #nilai rapat energi

#Nilai x dan y interpolasi 
xnew = np.linspace(x[0], x[-1], num = 50)
ynew = []

#Menentukan pangkat
m = len(x)
n = m-1

#Code
print('-----------------------------------------------------------')
print('| {0:^10} | {1:^18} | {2:^21} | '.format('N0','\u03BB','U(\u03BB, 5000)'))
print('|------------|--------------------|-----------------------|')

for i in range (len(xnew)):
    yp = 0
    for j in range (n+1):
        p = 1
        for k in range (n+1):
            if j != k:
                p*= (xnew[i] - x[k])/(x[j] - x[k])
        yp += y[j] * p
    ynew.append(yp)

for i in range(len(xnew)):
    no+=1
    print('| {0:^10} | {1:^18} | {2:^21} | '.format(no,xnew[i],ynew[i]))
    
print('-----------------------------------------------------------')

#Rapat energi maksimum 
xmax = np.where(np.max(ynew) == ynew)
ymax = np.max(ynew)
lamda = xnew[5]

#Persamaan
pers = ((8*pi*h*c)/(lamda*10**-9))*(1/((e**((h*c)/(lamda*10**-9*kb*T)))-1)) 

#Mencetak
print("\n Maka rapat energi maksimum dengan interpolasi lagrange adalah ", ymax)
print("\n Maka rapat energi maksimum dengan persamaan rapat energi adalah ", pers)
print("\n Maka jilai lamda yang berkolerasi adalah ", lamda)
```

### 8. Source Code
**Metode Interpolasi Newton**

```python
import numpy as np
import matplotlib.pyplot as plt
import math

#Diketahui
h = 6.626*10**-34
c = 2.998*10**8
kb = 1.381*10**-23
pi = np.pi
T = 5000
e = np.e
no = 0

#Nilai x dan y
x = [297, 721, 972, 1020, 1434, 1726] #nilai lamda
y = [75865, 127919, 134231, 286420, 314502, 482845] #nilai rapat energi

#Nilai x dan y interpolasi 
xnew = np.linspace(x[0], x[-1], num = 50)
ynew = []

#Menentukan pangkat
m = len(x)
n = m-1

#Code
print('-----------------------------------------------------------')
print('| {0:^10} | {1:^18} | {2:^21} | '.format('N0','\u03BB','U(\u03BB, 5000)'))
print('|------------|--------------------|-----------------------|')

def coeffts(x,y):
    koef = []
    for i in range (1,n+2):
        koef.append(y[0])
        for j in range (1,n+2-i):
            y[j-1] = (y[j]-y[j-1])/(x[j+i-1]-x[j-1])
        y.pop()
    return koef
    
def evalPoly(a,xData,x):
    n = len(xData) - 1 
    p = a[n]
    for k in range(1,n+1):
        p = a[n-k] + (x -xData[n-k])*p
    return p

a = coeffts(x,y)
p = evalPoly(a,x,xnew)

for i in range (len(xnew)):
    ynew = p
    no+=1
    print('| {0:^10} | {1:^18} | {2:^21} | '.format(no,xnew[i],ynew[i]))
    
print('-----------------------------------------------------------')

#Rapat energi maksimum 
xmax = np.where(np.max(ynew) == ynew)
ymax = np.max(ynew)
lamda = xnew[5]

#Persamaan
pers = ((8*pi*h*c)/(lamda*10**-9))*(1/((e**((h*c)/(lamda*10**-9*kb*T)))-1)) 

#Mencetak
print("\n Maka rapat energi maksimum dengan interpolasi Newton adalah ", ymax)
print("\n Maka rapat energi maksimum dengan persamaan rapat energi adalah ", pers)
print("\n Maka nilai lamda yang berkolerasi adalah ", lamda)
```
