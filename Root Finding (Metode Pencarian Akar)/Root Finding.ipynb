### 1. Problem Statement

Menghitung nilai jarak tempuh peluru terhadap x dengan menggunakan metode Bisection dan Newton-Raphson.

### 2. Persamaaan Matematika

* Fungsi ketinggian peluru terhadap x sebagai berikut:<br><br>
$Y(x) = x.tan(\theta) - \frac{1}{2}g(\frac{x}{v.cos(\theta)})^{2}$ <br><br>
* Diketahui:<br><br>
$V_{maks} = 800 m/s$<br>
$\theta = 45^{o}$<br>
$g = 9,8 m/s^{2}$

* Mencari nilai $X_{eksak}$ atau nilai sebenarnya dari fungsi di atas dengan persamaan sebagai berikut :<br><br>
$X_{eksak} = \frac{v^{2}}{g}sin(2\theta)$

* Nilai Error relatif dapat dicari dengan persamaan:<br><br>
$error = \left|\frac{x_{eksak}-x_{approx}}{x_{eksak}} \right|$

* Mencari nilai x dengan metode Bisection :<br><br>
  Jika $f(a)f(c) < 0$ , maka nilai b akan menjadi nilai c yang baru (b = c)
  
  Jika $f(a)f(c) > 0$, maka nilai a akan menjadi nilai c yang baru
  
  Jika $f(a)f(c) = 0$, maka c adalah akarnya

* Rumus metode Newton-Raphson :<br><br>
$x_{i+1} = x_{i} - \frac {f(x_{i})}{f'(x_{i})}$<br><br>

### 5. Source Code

* ## Metode Bisection

```python
import numpy as np
import matplotlib.pyplot as plt

a = 500
b = 90000
g = 9.8
v = 800
teta = np.pi/4
tan = np.tan(teta)
cos = np.cos(teta)
turn = 0
xeksak = ((v**2)/g)*np.sin(2*teta)


def f(x):
    f = (x*tan) - ((g/2)*((x**2)/((v*cos)**2)))
    return f

print('**Metode Bisection**')
print('|**********************************************************************************************************************|')
print('| {0:^10} | {1:^18} | {2:^18} | {3:^18} | {4:^18} | {5:>18} | '.format('Iterasi ke-','c','f(a).f(c)','a','b','Error'))
print('|*************|********************|********************|********************|********************|********************|')
while b-a > 0.0001:
    c = (a+b)/2
    error = round(abs((xeksak - c)/xeksak), 15)
    print('| {0:^11} | {1:^18} | {2:^18} | {3:^18} | {4:^18} | {5:>18} | '.format(turn, round(c,8), round(f(a)*f(c),6), round(a,8), round(b,8), error))
    turn += 1
    if f(a)*f(c) < 0:
        b = c
    else:
        a = c
print('|**********************************************************************************************************************|\n')
print ("Program Selesai")
```

* ## Metode Newton-Raphson

```python
import numpy as np
import matplotlib.pyplot as plt

a = 100000
g = 9.8
v = 800
teta = np.pi/4
tan = np.tan(teta)
sin = np.sin(teta)
cos = np.cos(teta)
turn = 0
iterasi = 100
xeksak = ((v**2)/g)*np.sin(2*teta)


def f(x):
    f = (x*tan) - ((g/2)*((x**2)/((v*cos)**2)))
    return f

def f2(x):
    f = (sin/cos)-((g*x)/((v**2)*(cos**2)))
    return f

print('**Metode Newton-Raphson**')
print('|****************************************************************|')
print('| {0:^10} | {1:^20} | {2:^25} | '.format('Iterasi ke-','Xapprox','Error'))
print('|*************|**********************|***************************|')

for i in range (0,iterasi) :
    c = a - (f(a)/f2(a))
    error = (abs((xeksak - c)/xeksak))
    print('| {0:^11} | {1:^20} | {2:^25} | '.format(turn, c,error))
    turn += 1
    if abs(f(a)) != 0:
        a=c
    else:
        break
print('|****************************************************************|\n')
print ("Program Selesai")
```
